```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\sachin\Python\Human Resource Management\output.csv")
data
```

```python
data.dtypes
```

```python
# Remove backticks (`) and convert to integer

data["State Code"] = data["State Code"].str.replace("`", "").astype(int)
data["District Code"] = data["District Code"].str.replace("`", "").astype(int)
data["Division"] = data["Division"].str.replace("`", "").astype(int)
data["Group"] = data["Group"].str.replace("`", "").astype(int)
data["Class"] = data["Class"].str.replace("`", "").astype(int)
```

### Clustering 

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# -----------------------------
# Pick the text column for NLP
# -----------------------------
TEXT_COL = "NIC Name"   # change to your text column name in 'data'

# -----------------------------
# TF-IDF Vectorization
# -----------------------------
vectorizer = TfidfVectorizer(stop_words="english", max_df=0.8, min_df=2)
X = vectorizer.fit_transform(data[TEXT_COL].astype(str))

# -----------------------------
# KMeans Clustering
# -----------------------------
n_clusters = 6   # tune this (try 4–10 for better grouping)
kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
data["Cluster"] = kmeans.fit_predict(X)

# -----------------------------
# Show top terms per cluster
# -----------------------------
terms = vectorizer.get_feature_names_out()
order_centroids = kmeans.cluster_centers_.argsort()[:, ::-1]

for i in range(n_clusters):
    top_terms = [terms[ind] for ind in order_centroids[i, :10]]
    print(f"Cluster {i}: {', '.join(top_terms)}")

# -----------------------------
# Save results
# -----------------------------
data.to_csv(r"C:\sachin\Python\Human Resource Management\industry_clusters.csv", index=False)
print("Clustered data saved!")
```
